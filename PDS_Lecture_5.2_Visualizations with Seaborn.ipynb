```python
###
# To plot the scatterplot only for male, filter the data on male
###

sns.relplot(data = tips[tips.sex == "Male" ], x = "total_bill", y = "tip", hue="time")
```

### Categorical Plots

```python
tips.head()
```

```python
sns.catplot(data = tips, x = "sex", y = "total_bill")
plt.show()
```

```python
sns.catplot(data = tips, x = "sex", y = "total_bill", jitter=False)
plt.show()
```

```python
# Set the palette using the name of a palette:
sns.set_palette("dark")

sns.catplot(data = tips, x = "sex", y = "total_bill", kind="swarm")
plt.show()
```

```python
# Save a palette to a variable:
palette = sns.color_palette("bright")

# Use palplot and pass in the variable:
sns.palplot(palette)

# Set the palette using the name of a palette:
sns.set_palette("bright")

sns.catplot(data = tips, x = "sex", y = "total_bill", kind="swarm", hue="time")
plt.show()
```

```python
sns.set_palette("bright")
sns.catplot(data = tips, x = "sex", y = "total_bill", kind="box")
```

```python
sns.catplot(data = tips, x = "sex", y = "total_bill", hue="time", kind="box")
```

```python
sns.catplot(data = tips, x = "sex", y = "total_bill", hue="time", kind="violin",
           palette= "ch:r=.25,l=.85")
```

```python
###
# numpy.linspace(start, stop, num = 50, endpoint = True, retstep = False, dtype = None) 
# Returns an array of number spaces evenly w.r.t interval. 
# Similar to arange but instead of step it uses sample number.

# Parameters :

# -> start  : [optional] start of interval range. By default start = 0
# -> stop   : end of interval range (inclusive)
# -> restep : If True, return (samples, step). By deflut restep = False
# -> num    : [int, optional] No. of samples to generate
# -> dtype  : type of output array

def sinplot():
    x = np.linspace(0, 14, 100)
    for i in range(1,7): # Draw seven plots
        plt.plot(x, np.sin(x + i*0.5) * (7-i))
```

```python
sinplot()
```

```python
###
# sns.set() resets everything to seaborn style

# When we execute the above function, everything remains the same
# but the plot is displayed in seaborn style
###
sns.set()
sinplot()
```
